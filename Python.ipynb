# Funções e Pandas

## Funções

```python
list_dicio = [{'fruta':'banana', 'quantidade':40, 'cor':'amarelo', 'preço':4.00},
 {'fruta':'uva', 'quantidade':10, 'cor':'roxo', 'preço':8.00},
 {'fruta':'pessego', 'quantidade':20, 'cor':'laranja', 'preço':10.00}]
 
for dicio in list_dicio:
  print(dicio['preço'])
  if dicio['preço'] > 8:
    print('maior do que 8')
  else:
    print('menor do que 8')
   
```

```python
def menor_igual_a_8(x):
  for dicio in x:
    print(dicio['preço'])
    if dicio['preço'] > 8:
      print('maior do que 8')
    else:
      print('menor do que 8')

### Def = definir função
### * A função já tem que vir com alguma variável na hora que você define ela, para a função entender o que é para fazer
### * As palavras têm que ser unidas por underline
```

```python
menor_igual_a_8(list_dicio)

### Na hora de usar a funçaõ, é só colocar o nome dela e, entre parêntesis, a variável que você quer.
```

```python
bolinha = [{'fruta':'banana', 'quantidade':40, 'cor':'amarelo', 'preço':6.00},
           {'fruta':'uva', 'quantidade':10, 'cor':'roxo', 'preço':7.00},
           {'fruta':'pessego', 'quantidade':20, 'cor':'laranja', 'preço':4.00}]

menor_igual_a_8(bolinha)
```

Exercício: Fazer um algoritmo que, ao receber o salário atual de um funcionário, calcule o valor do novo salário reajustado de acordo com a tabela abaixo:

image.png

```python
salario = 6000
if salario > 500:
  salario = salario*1.15 
elif 500 <= salario <= 1000:
  salario = salario*1.10
elif salario <= 1000:
  salario = salario*1.15
print(salario)
```

```python
def reajuste(salario):
  if salario > 500:
    salario = salario*1.15 
  elif 500 <= salario <= 1000:
    salario = salario*1.10
  elif salario <= 1000:
    salario = salario*1.15
  print("R$",round(salario,2))
  return(salario)

salario_reajustado = reajuste(6000)
```
